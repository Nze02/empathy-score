### why we used MLP modelling process here   

    - An MLP (Multilayer Perceptron) model was used for the empathy score prediction task because it is well-suited for solving regression problems. In this case, the goal was to predict the empathy score based on eye tracking data. The input features included various aspects of eye movements, such as gaze direction, pupil diameter, and eye position, among others.

  - An MLP model is a type of feedforward artificial neural network that is composed of multiple layers of interconnected nodes. The nodes in each layer are connected to the nodes in the next layer, allowing information to flow forward through the network. This architecture is suitable for non-linear regression problems, making it a good fit for the empathy score prediction task.

  - Additionally, MLP models are capable of learning complex relationships between the input features and target output. By training on a large dataset, the model can learn to make accurate predictions based on the relationships between the features and target. In this case, the model was trained on a dataset of eye tracking data, which allowed it to learn the relationship between eye movements and empathy scores.

  - Furthermore, MLP models are highly customizable, allowing for the creation of models with a wide range of architectures. In this case, multiple layers with different numbers of nodes were used to allow for the modeling of complex relationships between the input features and target.

  - In conclusion, the MLP model was an appropriate choice for this task due to its ability to model non-linear relationships, its ability to learn complex relationships between features and target, and its customizability. The use of this model has allowed for the accurate prediction of empathy scores based on eye tracking data.
